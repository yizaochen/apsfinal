```python
import matplotlib.pyplot as plt
import numpy
```

```python
x_array = range(5)
y_array = [1.119, 0.948, 2.644, 3.821, 4.522]
xticks = ['gender', 'Residence_type', 'ever_married', 'hypertension', 'heart_disease']
```

```python
fig, ax = plt.subplots(figsize=(6,6), facecolor="white")

ax.bar(x_array, y_array, 1, edgecolor="white")
ax.set_xticks(x_array)
ax.set_xticklabels(xticks)
ax.axhline(1, color="red", alpha=0.8)
ax.set_ylabel("Odds Ratio")
ax.tick_params(axis="x", labelrotation=30)

plt.tight_layout()
plt.savefig("oddsratio_bar.png", dpi=200, transparent=False)
plt.show()
```
