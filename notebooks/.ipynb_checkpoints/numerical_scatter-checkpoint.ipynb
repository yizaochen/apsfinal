```python
import pandas as pd
from os import path
import matplotlib.pyplot as plt
from scipy import stats
```

### Part 1: Read Data

```python
csv_in =  path.join('..', 'data', 'healthcare-dataset-stroke-data.csv')
df_in = pd.read_csv(csv_in)
df_in.head()
```

### Part 2: Data clean
##### because bmi have NaN values

```python
mask = pd.isna(df_in["bmi"])
df_clean = df_in[~mask]
```

```python
def get_xy_stroke(df0, sele_key1, sele_key2):
    df_nostroke = df0[df0["stroke"] == 0]
    df_stroke = df0[df0["stroke"] == 1]
    return df_nostroke[sele_key1], df_nostroke[sele_key2], df_stroke[sele_key1], df_stroke[sele_key2]
```

### Part 2:

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,4))

ax = axes[0]
key1 = "bmi"
key2 = "age"
xarray_nostroke, yarray_nostroke, xarray_stroke, yarray_stroke = get_xy_stroke(df_clean, key1, key2)
ax.scatter(xarray_nostroke, yarray_nostroke, color="grey", label="no-stroke")
ax.scatter(xarray_stroke, yarray_stroke, color="red", label="stroke")
ax.set_xlabel(key1)
ax.set_ylabel(key2)
ax.legend()

ax = axes[1]
key1 = "avg_glucose_level"
key2 = "age"
xarray_nostroke, yarray_nostroke, xarray_stroke, yarray_stroke = get_xy_stroke(df_clean, key1, key2)
ax.scatter(xarray_nostroke, yarray_nostroke, color="grey", label="no-stroke")
ax.scatter(xarray_stroke, yarray_stroke, color="red", label="stroke")
ax.set_xlabel(key1)
ax.set_ylabel(key2)

ax = axes[2]
key1 = "bmi"
key2 = "avg_glucose_level"
xarray_nostroke, yarray_nostroke, xarray_stroke, yarray_stroke = get_xy_stroke(df_clean, key1, key2)
ax.scatter(xarray_nostroke, yarray_nostroke, color="grey", label="no-stroke")
ax.scatter(xarray_stroke, yarray_stroke, color="red", label="stroke")
ax.set_xlabel(key1)
ax.set_ylabel(key2)

plt.tight_layout()
#plt.savefig("numerical_scatter.png", dpi=200)
plt.show()
```
