```python
import pandas as pd
from os import path
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_in =  path.join('..', 'data', 'healthcare-dataset-stroke-data-cleanbmi.csv')
df_in = pd.read_csv(csv_in)
df_in.head()
```

### Filter Marry

```python
mask = (df_in['ever_married'] == 'Yes') & (df_in['stroke'] == 1)
df_stroke_marry = df_in[mask]
df_stroke_marry = df_stroke_marry.sort_values(by=['age'])
df_stroke_marry.head()
```

```python
# Get the frequency, PDF and CDF for each value in the series

# Frequency
stats_df = df_stroke_marry \
.groupby('age') \
['age'] \
.agg('count') \
.pipe(pd.DataFrame) \
.rename(columns = {'age': 'frequency'})

# PDF
stats_df['pdf'] = stats_df['frequency'] / sum(stats_df['frequency'])

# CDF
stats_df['cdf'] = stats_df['pdf'].cumsum()
stats_df = stats_df.reset_index()
stats_df_marry = stats_df
stats_df_marry.head()
```

### Filter No-Marry

```python
mask = (df_in['ever_married'] == 'No') & (df_in['stroke'] == 1)
df_stroke_nomarry = df_in[mask]
df_stroke_nomarry = df_stroke_nomarry.sort_values(by=['age'])
df_stroke_nomarry.head()
```

```python
# Get the frequency, PDF and CDF for each value in the series

# Frequency
stats_df = df_stroke_nomarry \
.groupby('age') \
['age'] \
.agg('count') \
.pipe(pd.DataFrame) \
.rename(columns = {'age': 'frequency'})

# PDF
stats_df['pdf'] = stats_df['frequency'] / sum(stats_df['frequency'])

# CDF
stats_df['cdf'] = stats_df['pdf'].cumsum()
stats_df = stats_df.reset_index()
stats_df_nomarry = stats_df
stats_df_nomarry.head()
```

```python
fig, ax = plt.subplots(figsize=(8,4), facecolor="white")

ax.plot(stats_df_marry['age'], stats_df_marry['cdf'], linewidth=2, color="red", label="Ever Marry")
ax.plot(stats_df_nomarry['age'], stats_df_nomarry['cdf'], linewidth=2, color="blue", label="Never Marry")
ax.legend()
ax.set_ylabel("Pr(Age $\leq$ x| Stroke, Marry-Status)", fontsize=14)
ax.set_xlabel("x (age)", fontsize=14)
ax.set_title("Interaction effect between Marry and Age")

plt.tight_layout()
plt.savefig("interaction_marry_age_test.png", dpi=200, transparent=False)
plt.show()
```
